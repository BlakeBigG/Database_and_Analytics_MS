### Recall the general structure of the PostgreSQL function.

```
CREATE FUNCTION function_name(... arguments and data types ...)
RETURNS return_datatype
as
$$
DECLARE
   ... function local variables ... ;
BEGIN
   ... body of logic ...

   RETURN result_value;
END;
$$
LANGUAGE plpgsql;
```

See full documentation at: http://www.postgresql.cn/docs/10/sql-createfunction.html

## 1  Create the `totalSites` function in your database schema.

1. Update the below function, then use the terminal to define it.
2. Then, show the result of the `\df` command for your function. 
3. Finally, run the function with the `SELECT` command.

**Step 1: Create function**

````python
%%sql

CREATE OR REPLACE FUNCTION totalSites()
RETURNS integer AS 
$$
DECLARE
    total integer;   -- define a variable total with a data type
BEGIN
   SELECT count(*) INTO total FROM site;   -- Note INTO operator
   RETURN total;
END;
$$ 
LANGUAGE plpgsql;
````

**Step 2: Probe information_schema.routines catalog to verify the function**

````python
%%sql

-- select * 
select routine_schema, routine_name, routine_type, routine_definition
from information_schema.routines
where routine_name = 'totalsites'
````

Or We can use psql to see the definition of the function

**Step 3: Show the result of running the function**

````python
%%sql

SELECT totalSites()
````

## 2  Create the `totalReadingsForSite` function in your database schema.

1. Update the below function, then use the terminal to define it.
2. Then, show the result of the `\df` command for your function. 
3. Finally, run the function with the `SELECT` command.

### Note: We are now using a parameter variable in the function call!

#### Note: how the `p_site` variable is used within the SQL `SELECT ... INTO ....`

##### Think about how you expect the calling of the function to change since it has a parameter.

**Warning, if the function is called wrong, you may see:**
```
ERROR:  function totalreadingsforsite() does not exist
LINE 1: select totalreadingsforsite();
               ^
HINT:  No function matches the given name and argument types. You might need to add explicit type casts.
```

**Step 1: Create function**

````python
%%sql

CREATE OR REPLACE FUNCTION totalReadingsForSite(p_site text)
RETURNS integer AS 
$$
DECLARE
    total integer;   -- define a variable total with a data type
BEGIN
    SELECT count(*) INTO total
    FROM visited v JOIN survey s ON (v.id::int=s.taken::int)
    WHERE v.site = p_site;
    RETURN total;
END;
$$ 
LANGUAGE plpgsql;
````

**Step 2: Probe information_schema.routines catalog to verify the function**

````python
%%sql

SELECT routine_schema, routine_name, routine_type, routine_definition
FROM information_schema.routines
WHERE routine_name ilike 'totalReadingsForSite'
````

Or We can use psql to see the definition of the function
